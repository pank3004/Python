### Pie chart
#### to understand the percentage distribution of the data

```python
langs=['C', 'Python', 'Java']
students=[20, 100, 40]

plt.pie(students, labels=langs)
plt.title('pie char')
plt.show()
```

```python
langs=['C', 'Python', 'Java']
students=[20, 100, 40]

plt.pie(students, labels=langs, autopct='%1.1f%%')
plt.title('pie char')
plt.show()
```

```python
explode=(0.0, 0.1, 0.2)

langs=['C', 'Python', 'Java']
students=[20, 100, 40]

plt.pie(students, labels=langs, autopct='%1.1f%%', explode=explode)
plt.title('pie char')
plt.show()
```

```python
explode=(0.0, 0.1, 0.2)

langs=['C', 'Python', 'Java']
students=[20, 100, 40]

plt.pie(students, labels=langs, autopct='%1.1f%%', explode=explode, shadow=True)
plt.title('pie char')
plt.show()
```
