#### 1. relplot 

```python
# Numerical data ploting
# relplot >> relationship between two variable
            # which is short for relationship plot. relplot takes some of the same arguments as scatterplot , such as data, x, y , and hue , but adds other arguments as well.
            # scatterplot() (with kind="scatter"; the default)
            # lineplot() (with kind="line")


sns.relplot(x='total_bill', y='tip', data=tips)  # by default kind='scatter
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='line')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker') 
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='size') 
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, style='size') 
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, style='smoker',hue='time') 
```

### 3 catplot()
##### This function provides access to several axes-level functions that show the relationship between a numerical and one or more categorical variables using one of several visual representations.
 #### The kind parameter selects the underlying axes-level function to use.

#### Categorical scatterplots:

stripplot() (with kind="strip"; the default)

swarmplot() (with kind="swarm")

####  Categorical distribution plots:

boxplot() (with kind="box")

violinplot() (with kind="violin")

boxenplot() (with kind="boxen")

#### Categorical estimate plots:

pointplot() (with kind="point")

barplot() (with kind="bar")

countplot() (with kind="count")
