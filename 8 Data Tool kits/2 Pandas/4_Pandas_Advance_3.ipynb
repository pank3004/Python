### 1. merge

```python
df1=pd.DataFrame({'key1': [1,2,4,5,6],
                  'key2': [4,5,6,7,8],
                  'key3':[3,4,5,6,6]})

df2 =  pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
}
)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2, how='inner')
```

```python
pd.merge(df1,df2)       # by default how='inner
```

```python
pd.merge(df1,df2, how='inner', on='key1') # bydefault on = comman vala
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1,df2, how='right')
```

```python
pd.merge(df1, df2, how='outer')  # all the keys from both dataframe
```

```python
df1, df2
```

```python
pd.merge(df1, df2, how='cross')
```

```python
df1, df2
```

```python
pd.merge(df1, df2, how='left', left_on='key2', right_on='key4')
```

### 2 join

```python
df1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                    index = ['a', 'b', 'c', 'd', 'e']
)

df2 =  pd.DataFrame({'key6':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
},
index = ['a', 'b', 'h', 'i', 'j']
)
```

```python
df1
```

```python
df2
```

```python
df1.join(df2, how='inner')
```

```python
df1.join(df2, how='left')
```

```python
df1.join(df2, how='right')
```

```python
df1.join(df2, how='outer')
```

```python
df1.join(df2, how='cross')
```

### merge() vs join()

Both join and merge can be used to combines two dataframes but the join method combines two dataframes on the basis of their indexes whereas the merge method is more versatile and allows us to specify columns beside the index to join on for both dataframes
