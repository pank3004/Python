```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(1,10,200)  #  Return evenly spaced numbers over a specified interval.
```

```python
y=np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'--r')               # -- r means red desh line
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
plt.show()  #Display all open figures.
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y,'--r')  
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
plt.show()  
```

```python

plt.figure(figsize=(8,4))
plt.grid()
plt.plot(x,y,'--r')  
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
#plt.show() 
```

```python
x=np.random.rand(50)            # generate random data
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
colours=np.random.rand(50)
sizes=600*np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes,alpha=.4)    # alpha control the intensity of the points
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
         # by default c=None
```

```python
plt.scatter(x,y,c='b',s=sizes,alpha=.4) 
```

```python

plt.scatter(x,y,c='#DD2DF9')      # colour code 
```

```python
plt.scatter(x,y,c=colours,s=sizes,alpha=.4)    # alpha control the intensity of the points
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
```

```python
plt.scatter(x,y,c=colours,s=sizes,alpha=.8,marker='x')    # alpha control the intensity of the points
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")


'''
'.': Point marker
',', 'o': Circle marker
'v', '^', '<', '>': Triangle markers pointing downwards, upwards, leftwards, and rightwards respectively
's': Square marker
'p': Pentagon marker
'h', 'H': Hexagon markers (small and large, respectively)
'*': Star marker
'+': Plus marker
'x': Cross marker
'D', 'd': Diamond markers (large and small, respectively)
'|', '_': Vertical and horizontal line markers, respectively
'''
```

```python
plt.bar(x,y)    # categorical vs numerical data plot in vertical
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```

```python
plt.barh(x,y)    # categorical vs numerical data plot int horijonatl
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```
